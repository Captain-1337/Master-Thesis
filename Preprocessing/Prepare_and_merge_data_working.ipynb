### Import Libraries

```python
import pandas as pd
dates_no_gaps = pd.read_excel(open('All_Values_RavenPack_FinBERT_CrudeBERT_CrudeOil.xlsx', 'rb'), sheet_name='Combined_without_gaps', engine='openpyxl')
prices_crude_oil = pd.read_excel(open('All_Values_RavenPack_FinBERT_CrudeBERT_CrudeOil.xlsx', 'rb'), sheet_name='Prices_Crude_Oil', engine='openpyxl')  
sentiment_scores_crudebert = pd.read_excel(open('All_Values_RavenPack_FinBERT_CrudeBERT_CrudeOil.xlsx', 'rb'), sheet_name='DS_Sentiment_Scores_CrudeBERT', engine='openpyxl')  
sentiment_scores_crudebertv2 = pd.read_excel(open('All_Values_RavenPack_FinBERT_CrudeBERT_CrudeOil.xlsx', 'rb'), sheet_name='DS_Sentiment_Scores_CrudeBERTv2', engine='openpyxl')
sentiment_scores_finbert = pd.read_excel(open('All_Values_RavenPack_FinBERT_CrudeBERT_CrudeOil.xlsx', 'rb'), sheet_name='DS_Sentiment_Scores_FinBERT', engine='openpyxl')
sentiment_scores_crudebertt4 = pd.read_excel(open('All_Values_RavenPack_FinBERT_CrudeBERT_CrudeOil.xlsx', 'rb'), sheet_name='DS_Sentiment_Scores_CrudeBERTT4', engine='openpyxl')
google_trends_web = pd.read_excel(open('All_Values_RavenPack_FinBERT_CrudeBERT_CrudeOil.xlsx', 'rb'), sheet_name='Google_Trends_Web', engine='openpyxl')
```

### Reading and setting date without gaps as index

```python
dates_no_gaps['Date']=pd.to_datetime(dates_no_gaps['Date'])
dates_no_gaps.set_index('Date',inplace=True)
dates_no_gaps.head()
```

### Reading values of CrudeBERT

```python
#sentiment_scores_crudebert['Date']=pd.to_datetime(sentiment_scores_crudebert['Date'])
sentiment_scores_crudebert.set_index('Date',inplace=True)
sentiment_scores_crudebert.head()
```

### Reading values of CrudeBERTv2

```python
#sentiment_scores_crudebert['Date']=pd.to_datetime(sentiment_scores_crudebert['Date'])
sentiment_scores_crudebertv2.set_index('Date',inplace=True)
sentiment_scores_crudebertv2.head()
```

### Reading values of CrudeBERT_T4

```python
#sentiment_scores_crudebert['Date']=pd.to_datetime(sentiment_scores_crudebert['Date'])
sentiment_scores_crudebertt4.set_index('Date',inplace=True)
sentiment_scores_crudebertt4.head()
```

### Reading values of FinBERT

```python
sentiment_scores_finbert.set_index('Date',inplace=True)
sentiment_scores_finbert.head()
```

### Reading values of Google Trends

```python
google_trends_web.set_index('Date',inplace=True)
google_trends_web.head()
```

### Merging values of WTI Crude Oil with index

```python
price_crude_oil = prices_crude_oil[['Date' , 'Price']]
price_crude_oil['Date']=pd.to_datetime(price_crude_oil['Date'])
price_crude_oil.set_index('Date',inplace=True)
price_crude_oil.head()
```

### Merging all values with index

```python
combined_data = dates_no_gaps.join(sentiment_scores_crudebert, how='outer').join(sentiment_scores_crudebertv2, how='outer')\
.join(sentiment_scores_crudebertt4, how='outer')\
.join(sentiment_scores_finbert, how='outer')\
.join(price_crude_oil, how='outer').join(google_trends_web, how='outer')
```

```python
combined_data[40:50].head(10)
```

```python
combined_data = combined_data[1:]
```

### Interpolating (quadratic) gaps

```python
interpolated = combined_data.interpolate(method='quadratic')
```

### Saving Dataframe with gaps as Excel-Sheet

```python
combined_data.to_excel("quadratic_test_combined_data_ravenpack_crudebert_finbert_oil_googletrends_t4.xlsx")
```

### Saving interpolated Dataframe without gaps as Excel-Sheet

```python
interpolated.to_excel("quadratic_test_interpolated_ravenpack_crudebert_finbert_oil_googletrends_t4.xlsx")
```
